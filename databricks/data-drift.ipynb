**create a data drift monitor**

```
from azureml.core import Workspace, Dataset
from azureml.datadrift import DataDriftDetector, AlertConfiguration

alert_email = AlertConfiguration(['chiku@microsoft.com']) # replace with your email to recieve alerts from the scheduled pipeline after enabling

monitor = DataDriftDetector.create_from_datasets(workspace, 'drift-monitor-1',  baseline,  new,
                                                      compute_target='dev-cluster', 
                                                      frequency='Week', 
                                                      feature_list=None, 
                                                      drift_threshold=0.1,  # threshold from 0 to 1 for email alerting
                                                      latency=0, # SLA in hours for target data to arrive in the dataset
                                                      alert_config=alert_email #alert_configuration=alert_email # email addresses to send alert)
                                                )
```

```
from azureml.core import Workspace, Dataset
from azureml.datadrift import DataDriftDetector, AlertConfiguration

# get data drift detector by name
monitor = DataDriftDetector.get_by_name(workspace, 'drift-monitor-1')
```

```
# set up feature list
features = ['Station', 'Code', 'Bound', 'Line', 'Vehicle']
# update data drift detector
monitor = monitor.update(feature_list=features)
```

```
# run a backfill for May through today
import datetime as dt
backfill1 = monitor.backfill(datetime(2019, 10, 1), datetime(2020, 1, 1))
```

```
# disable the pipeline schedule for the data drift detector
monitor = monitor.disable_schedule()

# enable the pipeline schedule for the data drift detector
monitor = monitor.enable_schedule()
```
